```python
import numpy as np
import matplotlib.pyplot as plt
from einops import rearrange
```

# 2.a

```python
X=np.linspace(-2,2,1000)
xi_const=[i for i in np.linspace(0,10,20)]
eta_const=[i for i in np.linspace(0,8,20)]
for i, xi in enumerate(xi_const):
    transformed_y = ((xi-X**2/2)/2)**2
    label = '$\\xi$ = const' if i == 0 else None
    plt.plot(X, transformed_y,color='red', label=label)
for i, eta in enumerate(eta_const):
    transformed_y = ((eta+X**2/2)/2)**2
    label = '$\\eta$ = const' if i == 0 else None
    plt.plot(X, transformed_y, color='blue',label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-2,2])
plt.ylim([0,4])
plt.grid()
plt.legend()
plt.savefig('2a.png')
plt.show()
```

# 2.b


```python
X=np.linspace(-2,2,1000)
Y=np.linspace(-2,2,1000)
xi_const=[i for i in np.linspace(-10,10,20)]
eta_const=[i for i in np.linspace(-2,2,10)]
for i, xi in enumerate(xi_const):
    transformed_X = -np.log(np.exp(-Y)+xi)
    label = '$\\xi$ = const' if i == 0 else None
    plt.plot(transformed_X, Y,color='red', label=label)
for i, eta in enumerate(eta_const):
    transformed_y = transformed_X-transformed_X+eta
    label = '$\\eta$ = const' if i == 0 else None
    plt.plot(X, transformed_y, color='blue',label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()
plt.legend()
plt.savefig('2b.png')
plt.show()
```

## 2.c

```python
X=np.linspace(-2,2,1000)
Y=np.linspace(-2,2,1000)
xi_const=[i for i in np.linspace(-10,10,15)]
eta_const=[i for i in np.linspace(-10,10,15)]
for i, xi in enumerate(xi_const):
    transformed_Y = xi-3*X
    label = '$\\xi$ = const' if i == 0 else None
    plt.plot(X, transformed_Y,color='red', label=label)
for i, eta in enumerate(eta_const):
    transformed_y = X+eta
    label = '$\\eta$ = const' if i == 0 else None
    plt.plot(X, transformed_y, color='blue',label=label)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([-2,2])
plt.ylim([-2,2])
plt.grid()
plt.legend()
plt.savefig('2c.png')
plt.show()
```
