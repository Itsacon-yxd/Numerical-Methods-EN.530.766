## 1 

By the chain rule differentiation, the following can be obtained:
$$
\Phi_{xx}=\frac{\partial}{\partial x}(\frac{\partial \Phi}{\partial x})=\frac{\partial}{\partial x}(\frac{\partial \Phi}{\partial \xi}\frac{\partial \xi}{\partial x}+\frac{\partial \Phi}{\partial \eta}\frac{\partial \eta}{\partial x})
$$
$$
=
\frac{\partial^2 \Phi}{\partial \xi^2}(\frac{\partial \xi}{\partial x})^2+\frac{\partial^2 \Phi}{\partial\xi\eta}\frac{\partial \eta}{\partial x}\frac{\partial \xi}{\partial x}+\frac{\partial \Phi}{\partial \xi}\frac{\partial^2 \xi}{\partial x^2}
+
\frac{\partial^2 \Phi}{\partial \eta^2}(\frac{\partial \eta}{\partial x})^2+\frac{\partial^2 \Phi}{\partial\eta\xi}\frac{\partial \xi}{\partial x}\frac{\partial \eta}{\partial x}+\frac{\partial \Phi}{\partial \eta}\frac{\partial^2 \eta}{\partial x^2}
$$
$$
=\Phi_{\xi\xi}\xi_x^2+\Phi_{\xi\eta}\eta_x\xi_x+\Phi_\xi\xi_{xx}+\Phi_{\eta\eta}\eta_x^2+\Phi_{\eta\xi}\xi_x\eta_x+\Phi_\eta\eta_{xx}
$$
$$
=\Phi_{\xi\xi}\xi_x^2+2\Phi_{\xi\eta}\eta_x\xi_x+\Phi_{\eta\eta}\eta_x^2+\Phi_\xi\xi_{xx}+\Phi_\eta\eta_{xx}
$$
Similarly, the following can be obtained for $\Phi_{yy}$:
$$
\Phi_{yy}=\Phi_{\xi\xi}\xi_y^2+2\Phi_{\xi\eta}\eta_y\xi_y+\Phi_{\eta\eta}\eta_y^2+\Phi_\xi\xi_{yy}+\Phi_\eta\eta_{yy}
$$
By the chain rule differentiation, the following can be obtained for $\Phi_{xy}$:
$$
\Phi_{xy}=\frac{\partial}{\partial x}(\frac{\partial \Phi}{\partial y})=\frac{\partial}{\partial x}(\Phi_\xi\xi_y+\Phi_\eta\eta_y)
$$
$$
=\Phi_{\xi\xi}\xi_x\xi_y+\Phi_{\xi\eta}\eta_x\xi_y+\Phi_\xi\xi_{xy}+\Phi_{\eta\xi}\xi_x\eta_y+\Phi_{\eta\eta}\eta_x\eta_y+\Phi_\eta\eta_{xy}
$$
Plugging the above results back into the PDE, the following can be obtained:
$$
(a\xi_x^2+b\xi_x\xi_y+c\xi_y^2)\Phi_{\xi\xi}+[2a\xi_x\eta_x+b(\xi_x\eta_y+\xi_y\eta_x)+2c\xi_y\eta_y]\Phi_{\xi\eta}+(a\eta_x^2+b\eta_x\eta_y+c\eta_y^2)\Phi_{\eta\eta}=
$$
$$
g(x,y)-(a\xi_{xx}+b\xi_{xy}+c\xi_{yy})\Phi_\xi-(a\eta_{xx}+b\eta_{xy}+c\eta_{yy})\Phi_\eta-f\Phi
$$
Therefore:
$$
A=a\xi_x^2+b\xi_x\xi_y+c\xi_y^2
$$
$$
B=2a\xi_x\eta_x+b(\xi_x\eta_y+\xi_y\eta_x)+2c\xi_y\eta_y
$$
$$
C=a\eta_x^2+b\eta_x\eta_y+c\eta_y^2
$$
$$
H=g(x,y)-(a\xi_{xx}+b\xi_{xy}+c\xi_{yy})\Phi_\xi-(a\eta_{xx}+b\eta_{xy}+c\eta_{yy})\Phi_\eta-f\Phi
$$
