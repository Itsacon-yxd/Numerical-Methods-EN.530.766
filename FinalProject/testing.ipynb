```python
from jax_solver import harmonic_solution
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
from jax_solver import Pt_GS_Solver, get_accuracy
```

```python
solver = Pt_GS_Solver(num_pts=160, boundary_list=jnp.array([[0.5, 0.5, 0.25]]))
solution, residual = solver.solve(w=1.8, max_iter=5000)
```

```python
plt.plot(residual)
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('Residual')
plt.title('Convergence of Point Gauss-Seidel Solver')
plt.grid()
plt.show()
```
